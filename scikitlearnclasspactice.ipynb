```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
X, y = load_iris(return_X_y=True)
```

```python
X,y
```

```python
#Find optimum number of cluster
sse = [] #SUM OF SQUARED ERROR
for k in range(1,11):
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
sns.set_style("whitegrid")
g=sns.lineplot(x=range(1,11), y=sse)

g.set(xlabel ="Number of cluster (k)",
      ylabel = "Sum Squared Error",
      title ='Elbow Method')

plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 2)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
pred = kmeans.fit_predict(X)
pred
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c = pred, cmap=cm.Accent)
plt.grid(True)
for center in kmeans.cluster_centers_:
    center = center[:2]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")

plt.subplot(1,2,2)
plt.scatter(X[:,2],X[:,3],c = pred, cmap=cm.Accent)
plt.grid(True)
for center in kmeans.cluster_centers_:
    center = center[2:4]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Step 1: Generate a synthetic dataset
X, y = make_classification(
    n_samples=1000,  # 1000 samples
    n_features=20,   # 20 features
    n_informative=15,  # 15 features are informative
    n_redundant=5,   # 5 features are redundant
    random_state=42, # Seed for reproducibility
    class_sep=1.5,   # Better separation of classes
)

# Step 2: Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train Naive Bayes Model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_predictions = nb_model.predict(X_test)

# Step 4: Train SVM Model
svm_model = SVC(kernel='linear', random_state=42)  # Using a linear kernel for simplicity
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)

# Step 5: Evaluate Models
nb_accuracy = accuracy_score(y_test, nb_predictions)
svm_accuracy = accuracy_score(y_test, svm_predictions)

nb_report = classification_report(y_test, nb_predictions)
svm_report = classification_report(y_test, svm_predictions)

(nb_accuracy, svm_accuracy, nb_report, svm_report)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay

# Step 1: Generate a synthetic dataset
X, y = make_classification(
    n_samples=1000,
    n_features=2,  # Reduced to 2 features for visualization
    n_informative=2,
    n_redundant=0,
    random_state=42,
    class_sep=1.5
)

# Step 2: Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train Naive Bayes Model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
nb_predictions = nb_model.predict(X_test)

# Train SVM Model
svm_model = SVC(kernel='linear', random_state=42, probability=True)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)

# Step 4: Evaluate Models
nb_accuracy = accuracy_score(y_test, nb_predictions)
svm_accuracy = accuracy_score(y_test, svm_predictions)

print(f"Naive Bayes Accuracy: {nb_accuracy * 100:.2f}%")
print(f"SVM Accuracy: {svm_accuracy * 100:.2f}%")

# Classification Reports
print("\nNaive Bayes Classification Report:")
print(classification_report(y_test, nb_predictions))

print("\nSVM Classification Report:")
print(classification_report(y_test, svm_predictions))

# Step 5: Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
ConfusionMatrixDisplay.from_estimator(nb_model, X_test, y_test, ax=axes[0], cmap="Blues")
axes[0].set_title("Naive Bayes Confusion Matrix")
ConfusionMatrixDisplay.from_estimator(svm_model, X_test, y_test, ax=axes[1], cmap="Blues")
axes[1].set_title("SVM Confusion Matrix")
plt.tight_layout()
plt.show()

# Step 6: Decision Boundaries
def plot_decision_boundary(model, X, y, title, ax):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    ax.set_title(title)
    ax.legend(*scatter.legend_elements(), title="Classes")

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
plot_decision_boundary(nb_model, X_test, y_test, "Naive Bayes Decision Boundary", axes[0])
plot_decision_boundary(svm_model, X_test, y_test, "SVM Decision Boundary", axes[1])
plt.tight_layout()
plt.show()
```
